```python
import urllib.request
import json
import time
import folium

# Set up the world map
w_map = folium.Map(zoom_start=5, tiles='Stamen Terrain', attr='Map data &copy; <a href="https://www.mapbox.com/">Mapbox</a>')


# Set up the producer link
prod_url = "http://api.open-notify.org/iss-now.json"

# Set the duration of the producer in seconds
prod_time = 3600

# Set the streaming interval in seconds
stream_interval = 5

# Set the start time of the producer
initial_time = time.time()

# Start the producer
while time.time() - initial_time < prod_time:
    # Get the data from the producer
    response = urllib.request.urlopen(prod_url)
    sat_data = response.read().decode("utf-8")
    sat_data = json.loads(sat_data)

    # Get the latitude and longitude from the data
    lat = float(sat_data["iss_position"]["latitude"])
    long = float(sat_data["iss_position"]["longitude"])

    # Add the satellite location to the world map
    folium.Marker([lat, long]).add_to(w_map)

    # Wait for the streaming interval before fetching data again
    time.sleep(stream_interval)
#display world map
display(w_map)
```
